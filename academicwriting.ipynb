```python
import pandas as pd
```

```python
df = pd.read_csv('../quant/alldata.csv')
df
```

### Keep only columns needed

```python
df = df[['Stkcd', 'Trddt', 'Adjprcwd']]
df.columns = ['c', 'd', 'p']
df
```

### Convert date format

```python
df['d'] = pd.to_datetime(df['d'], format='%Y-%m-%d')
df
```

```python
QUARTERS = ['2009-12-31']
for year in range(2010, 2022):
    for quarter in ['-3-31', '-6-30', '-9-30', '-12-31']:
        QUARTERS.append(f'{year}{quarter}')
QUARTERS
```

```python
all_res_cols = []
for i in range(3,15,3):
    for j in ['-buy', '-sell', '-buy-sell']:
        all_res_cols.append(f'{i}{j}')
all_res_cols
```

```python
def rankByRisePercent(df):
    """
    Ranks the stocks by the percentage risen over the entire period.
    Returns a dataframe
    """
    df_res = calcRisePercent(df)
    df_res['rank'] = pd.qcut(df_res['rise_percent'], 10, labels = False)
    return df_res

def calcRisePercent(df):
    """
    Calculates the percentage that the stock price has risen from the start of the period to the
    end. Returns a series
    """
    groupedBy = df.groupby('c')
    idx1 = groupedBy['d'].transform(min) == df['d']
    start_price = df[idx1][['c', 'p']].reset_index(drop=True).rename(columns={'p':'start_price'})
    idx2 = groupedBy['d'].transform(max) == df['d']
    end_price = df[idx2][['c', 'p']].reset_index(drop=True).rename(columns={'p':'end_price'})
    df = start_price.merge(end_price, on='c', how='inner')
    df['rise_percent'] = (df.end_price - df.start_price) / df.start_price
    return df

def getResults(df1, df2):
    df_res1 = rankByRisePercent(df1)
    df_res2 = calcRisePercent(df2)
    df_final_res = df_res2[['c', 'rise_percent']].merge(df_res1[['c', 'rank']], on='c', how='inner')
    return df_final_res.groupby('rank').mean()

def getFinalResults(df, j, k):
    df_final_res = pd.DataFrame(columns=['buy', 'sell', 'buy-sell'])
    for i in range(j, len(QUARTERS)-k):
        df_lookback = df[(df['d'] <= pd.to_datetime(QUARTERS[i], format='%Y-%m-%d')) &\
                         (df['d'] >= pd.to_datetime(QUARTERS[i-j], format='%Y-%m-%d') + pd.Timedelta(days=1))]
        df_forward = df[(df['d'] <= pd.to_datetime(QUARTERS[i+k], format='%Y-%m-%d')) &\
                        (df['d'] >= pd.to_datetime(QUARTERS[i], format='%Y-%m-%d') + pd.Timedelta(days=1))]
        df_res = getResults(df_lookback, df_forward)
        df_final_res.loc[len(df_final_res.index)] = [df_res.iloc[-1,1]/3, df_res.iloc[0,1]/3, \
                                                     df_res.iloc[-1,1]/3-df_res.iloc[0,1]/3]
    return df_final_res
```

```python
def getAllResults(df):
    df_all_res = pd.DataFrame(columns=['J'] + all_res_cols)
    for i in range(1,5):
        temp = [i*3]
        for j in range(1,5):
            print(i,j)
            temp += list(getFinalResults(df, i, j).mean())
        df_all_res.loc[len(df_all_res.index)] = temp
    return df_all_res
```
