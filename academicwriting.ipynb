```python
import pandas as pd
```

```python
df = pd.read_csv('../quant/alldata.csv')
df
```

### Keep only columns needed

```python
df = df[['Stkcd', 'Trddt', 'Adjprcwd']]
df.columns = ['c', 'd', 'p']
df
```

### Convert date format

```python
df['Trddt'] = pd.to_datetime(df['d'], format='%Y-%m-%d')
df
```

```python
df1q10 = df[df['d'] <= pd.to_datetime('2010-3-31', format='%Y-%m-%d')]
df1q10
```

```python
((df1q10.groupby('c').max() - df1q10.groupby('c').min())['p'] / df1q10.groupby('c').min()['p']).sort_values()
```

```python
def divideIntoGroups(df):
    groupedBy = df.groupby('c')
    res = (groupedBy.max() - groupedBy.min())['p'] / groupedBy.min()['p']
    return res.sort_values()

divideIntoGroups(df1q10)
```
